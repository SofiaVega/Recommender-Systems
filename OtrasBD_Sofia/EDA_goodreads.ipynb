```python
isbn_arr = df["isbn13"].values
print(isbn_arr)
```

```python
subjects = []
for isbn in isbn_arr:
  print("https://api2.isbndb.com/book/" + str(isbn))
  resp_isbn = req.get("https://api2.isbndb.com/book/" + str(isbn), headers=h)
  subjects.append(resp_isbn)
print(subjects)
```

```python
print(subjects)
```

```python
print(subjects[0].json().get("book").get("subjects"))
```

```python
f = open("demofile2.txt", "w")
for i in subjects:
  f.write(str(i.json()) + '\n')
f.close()
```

```python
s = []
forbiddenJson = {'message': 'Forbidden'}
for i in subjects:
  if i.json() == forbiddenJson:
    s.append([])
  else:
    print(i.json())
    s.append(i.json().get("book").get("subjects"))
print(s)
```

```python
subjects_series = pd.Series(s)
df["subjects"] = subjects_series.values
```

```python
df.to_csv(r'data_with_subject.csv', index = False)
```
