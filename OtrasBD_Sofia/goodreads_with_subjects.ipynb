```python
df = pd.read_csv('goodreads/data_with_subject.csv', error_bad_lines = False)
df.head()
print(df.groupby('subjects').size())
```

```python
print(df.groupby('subjects').count())
```

```python
def clean_alt_list(list_):
    print(type(list_))
    if list_ is None:
      return "[]"
    #print(list_)
    if type(list_) != type("str"):
      return "[]"
    #if list_ == "[]"
    #list_ = list_.replace(', ', '","')
    #list_ = list_.replace('[', '["')
    #list_ = list_.replace(']', '"]')
    #print(type(list_))
    #print(list_)
    return list_

def to_1D(series):
 return pd.Series([x for _list in series for x in _list])
```

```python
df["subjects"] = df["subjects"].apply(clean_alt_list)
df["subjects"] = df["subjects"].apply(eval)
to_1D(df["subjects"]).value_counts()
```

```python
subjects_series = to_1D(df["subjects"]).value_counts()
#print(subjects_series)
top10 = subjects_series.head(10)
```

```python
fig, ax = plt.subplots(figsize = (14,4))
ax.bar(top10.index,
        top10.values)
ax.set_ylabel("Frequency", size = 12)
ax.set_title("Book subjects", size = 14)
```
