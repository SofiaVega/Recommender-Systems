# Display Sample Records

```python
import gzip
import json
import re
import os
import sys
import numpy as np
import pandas as pd
```

**Specify your directory here:**

```python
DIR = './'
```

**This function shows how to load datasets**

```python
def load_data(file_name, head = 500):
    count = 0
    data = []
    with gzip.open(file_name) as fin:
        for l in fin:
            d = json.loads(l)
            count += 1
            data.append(d)
            
            # break if reaches the 100th line
            if (head is not None) and (count > head):
                break
    return data
```

**Load and display sample records of books/authors/works/series**

```python
interactions = load_data(os.path.join(DIR, 'goodreads_interactions_mystery_thriller_crime.json.gz'))
books = load_data(os.path.join(DIR, 'goodreads_books_mystery_thriller_crime.json.gz'))
reviews = load_data(os.path.join(DIR, 'goodreads_reviews_mystery_thriller_crime.json.gz'))
```

```python
import surprise
from surprise import KNNWithMeans
from surprise import Dataset
from surprise import Reader
```

```python
sim_options = {
    "name": "cosine",
    "user_based": False,  # Compute  similarities between items
}
algo = KNNWithMeans(sim_options=sim_options)
```

```python
reader = Reader(rating_scale=(1, 5))

# The columns must correspond to user id, item id and ratings (in that order).
data = Dataset.load_from_df(reviews_df[["user_id", "book_id", "rating"]], reader)
trainingSet = data.build_full_trainset()
```

```python
algo.fit(trainingSet)
```

```python
for index, row in reviews_subset.iterrows():
  print(row)
  print(algo.predict(row["user_id"], row["book_id"]))

algo.predict("8842281e1d1347389f2ab93d60773d4d", "6392944")
```
